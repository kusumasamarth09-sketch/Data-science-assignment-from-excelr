2.**ModeThe mode is the value that occurs most frequently in a dataset.**

```python
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv("/content/sales_data_with_discounts.csv")

# Display basic info
print(df.info())

# Identify numerical columns
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
print("Numerical Columns:", num_cols)

# Compute descriptive statistics
desc_stats = df[num_cols].describe().T
desc_stats['mode'] = [df[col].mode()[0] for col in num_cols]
print(desc_stats[['mean', '50%', 'mode', 'std']])
```

# **4.Boxplots**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
df = pd.read_csv("/content/sales_data_with_discounts.csv")
# Identify numerical columns
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()

for col in num_cols:
    plt.figure(figsize=(6, 4))
    plt.boxplot(df[col] )
    plt.title(f"Boxplot of {col}")
    plt.xlabel(col)
    plt.show()
```

# **5.Bar Chart Analysis for Categorical Columns**

```python
cat_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()
print("Categorical Columns:", cat_cols)

for col in cat_cols:
    plt.figure(figsize=(8, 4))
    df[col].value_counts().plot(kind='bar', color='skyblue')
    plt.title(f"Category Distribution: {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

# **6.Standardization of Numerical Variables**

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled = df.copy()
df_scaled[num_cols] = scaler.fit_transform(df[num_cols])

print("Before Standardization:")
print(df[num_cols].head())

print("\nAfter Standardization:")
print(df_scaled[num_cols].head())
```

```python
df[num_cols].hist(bins=20, figsize=(12, 8))
plt.suptitle("Before Standardization", fontsize=16)
plt.show()

df_scaled[num_cols].hist(bins=20, figsize=(12, 8))
plt.suptitle("After Standardization", fontsize=16)
plt.show()
```

# **6.Conversion of Categorical Variables into Dummy Variables**

```python
df_encoded = pd.get_dummies(df_scaled, columns=cat_cols, drop_first=True)
print(df_encoded.head())
```
