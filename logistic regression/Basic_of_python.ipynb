**Prime number**

```python

# checks a given number is prime or not.

num = int(input("Enter a number: "))

if num > 1:
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            print(f"{num} is not a prime number.")
            break
    else:
        print(f"{num} is a prime number.")
else:
    print("Prime numbers are greater than 1.")
```

**Product of Random Numbers**

```python

# This program generates two random numbers and asks the user to find their product.

import random

num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

print(f"What is the product of {num1} and {num2}?")
user_answer = int(input("Your answer: "))

if user_answer == num1 * num2:
    print("the correct answer.")
else:
    print("Incorrect. The correct answer is {num1 * num2}.")
```

**Squares of Even/Odd Numbers**

```python

# Choice: Even numbers (you can switch to odd by changing the condition)

print("Squares of even numbers between 100 and 200:")

for num in range(100, 201):
    if num % 2 == 0:  # Change to `num % 2 != 0` for odd numbers
        print(f"{num}^2 = {num ** 2}")
```

**Word Counter**

```python

# This program counts the frequency of each word in a given text.

input_text = "Python is fun. Python is powerful and Python is easy to learn!"
words = input_text.split()
word_count = {}

for word in words:
    word = word.lower()  # normalize to lowercase
    word_count[word] = word_count.get(word, 0) + 1

for word, count in word_count.items():
    print(f"'{word}': {count}")
```

**Check for Palindrome**

```python

# This function checks a string is a palindrome.

def is_palindrome(text):
    cleaned = ''.join(char.lower() for char in text if char.isalnum())  # remove spaces and punctuation
    return cleaned == cleaned[::-1]

# Example usage:
word = input("Enter a word or phrase: ")
if is_palindrome(word):
    print("It's a palindrome!")
else:
    print("It's not a palindrome.")
```
